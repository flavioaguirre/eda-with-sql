To analyze the data using SQL, you first need to load it into the SQLite database.
We'll create three tables as shown below:

1. **CENSUS_DATA**
2. **CHICAGO_PUBLIC_SCHOOLS**

As a first step we are going to create a Database called `chicago_data.db` and then establish a connection with it.

```python
%load_ext sql
```

```python
try: 
    conn = sqlite3.connect("../../database/chicago_data.db") 
    print("✅ Connection successful") 
except Exception as e: 
        print(f"❌ Error connecting:{e}")
```

We establish a connection between the `Magic SQL` module and the `chicago_data.db` database.

```python
%sql sqlite:///../../database/chicago_data.db
```

```python
# we read the csv files and save them in variables like Pandas DataFrames
census_data = pd.read_csv('../../data/raw/census_data.csv')
public_schools = pd.read_csv('../../data/raw/public_schools.csv')
crime_data = pd.read_csv('../../data/raw/crime_data.csv')

# We load the DataFrames into the database "FinalDB.db"
census_data.to_sql("chicago_socioeconomic_data", conn, if_exists='replace', index=False, method="multi")
crime_data.to_sql("chicago_crime_data", conn, if_exists='replace', index=False, method="multi")

# The insert parameters for the "CHICAGO_PUBLIC_SCHOOLS" query exceed
# The allowed values, so we load it in batches.

# We define the batch size to split the DataFrame into smaller parts.
batch_size = 500

# We iterate over the DataFrame in batches and load each batch into the SQL table.
for i in range(0, len(chicago_public_schools), batch_size):
    batch = chicago_public_schools.iloc[i:i+batch_size]
    batch.to_sql("chicago_public_schools_data", conn, if_exists='append', index=False)
```

We check if the tables were created correctly

Socioeconomic Indicators in Chicago:

```python
%sql SELECT * FROM chicago_socioeconomic_data LIMIT 5;
```

Chicago Public Schools:

```python
%sql SELECT * FROM chicago_public_schools_data LIMIT 5;
```

Chicago Crime Data:

```python
%sql SELECT * FROM chicago_crime_data LIMIT 5;
```

We close the connection

```python
conn.close()
print("Database connection closed.")
```
